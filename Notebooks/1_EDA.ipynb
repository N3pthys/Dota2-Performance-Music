```python
import numpy as np
# Bar Plot of Average GPM by Medal
plt.figure(figsize=(10, 6))
sns.barplot(x='Medal', y='GPM', data=data, estimator=np.mean)
plt.title('Average GPM by Medal')
plt.xlabel('Medal')
plt.ylabel('Average GPM')
plt.show()
```

```python
# Scatter Plot of GPM vs. XPM
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GPM', y='XPM', data=data, hue='WinOrLose', alpha=0.7)
plt.title('GPM vs. XPM')
plt.xlabel('Gold Per Minute (GPM)')
plt.ylabel('Experience Per Minute (XPM)')
plt.legend(title='Win or Lose')
plt.show()
```

```python
# Heatmap of Performance Metrics
plt.figure(figsize=(12, 8))
performance_metrics = data[['GPM', 'XPM', 'APM', 'Win Rate']]
sns.heatmap(performance_metrics.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Performance Metrics')
plt.show()
```

```python
# Scatter Plot of GPM vs. APM
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GPM', y='APM', hue='WinOrLose', data=df, alpha=0.6)
plt.title('GPM vs. APM')
plt.xlabel('Gold Per Minute (GPM)')
plt.ylabel('Actions Per Minute (APM)')
plt.legend(title='Win or Lose', loc='upper left')
plt.show()
```

```python
#  Pair Plot
sns.pairplot(data[['GPM', 'XPM', 'Final Net Worth', 'Win Rate']], diag_kind='kde')
plt.suptitle('Pair Plot of Performance Metrics', y=1.02)
plt.show()
```

```python
# Scatter Plot of Kills vs. Final Net Worth
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Kills', y='Final Net Worth', hue='WinOrLose', data=data, alpha=0.7)
plt.title('Kills vs. Final Net Worth')
plt.xlabel('Kills')
plt.ylabel('Final Net Worth')
plt.legend(title='Win or Lose')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Facet Grid of Kills by Music Playback
g = sns.FacetGrid(df, col='Music Listened', height=4, aspect=1.5)
g.map(sns.histplot, 'Kills', bins=10, kde=True)
g.set_titles(col_template="{col_name} Playback")
g.set_axis_labels('Kills', 'Frequency')
plt.show()


# Set up the figure for Deaths
g_deaths = sns.FacetGrid(df, col='Music Listened', height=4, aspect=1.5)
g_deaths.map(sns.histplot, 'Deaths', bins=10, kde=True)
g_deaths.set_titles(col_template="{col_name} Playback")
g_deaths.set_axis_labels('Deaths', 'Frequency')
plt.show()


# Set up the figure for Assists
g_assists = sns.FacetGrid(df, col='Music Listened', height=4, aspect=1.5)
g_assists.map(sns.histplot, 'Assists', bins=10, kde=True)
g_assists.set_titles(col_template="{col_name} Playback")
g_assists.set_axis_labels('Assists', 'Frequency')
plt.show()

```

```python
# Average GPM by Genre
plt.figure(figsize=(12, 6))
avg_gpm_by_genre = df.groupby('Genre')['GPM'].mean()
avg_gpm_by_genre.plot(kind='bar', color='skyblue')
plt.title('Average GPM by Genre')
plt.xlabel('Genre')
plt.ylabel('Average GPM')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Heatmap of Key Metrics
plt.figure(figsize=(12, 8))
correlation = df[['GPM', 'XPM', 'Kills', 'Deaths', 'Assists', 'Game Duration', 'Final Net Worth']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Key Metrics')
plt.show()
```
